```python
import requests
from bs4 import BeautifulSoup

def get_product_links(url):
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an error if the request fails

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all product link elements
    product_links = []
    for link in soup.find_all('a', class_='product-link'):  # Adjust the class name as needed
        href = link.get('href')
        if href and href.startswith('/'):  # Ensure it's a valid product link
            product_links.append(f"https://sunnamusk.com{href}")

    return product_links

# Example usage
url = "https://sunnamusk.com/collections/eau-de-parfum"
product_links = get_product_links(url)
print(product_links)
```

```python
from bs4 import BeautifulSoup

def extract_category_links(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    categories = {}

    # Find all main categories
    for sliderule_wrapper in soup.find_all("div", class_="sliderule__wrapper"):
        # Get the main category title
        main_category_button = sliderule_wrapper.find("button", class_="sliderow")
        if main_category_button:
            main_category_title = main_category_button.find("span", class_="sliderow__title")
            if main_category_title:
                main_category_name = main_category_title.get_text(strip=True)

                # Extract subcategories
                subcategories = []
                subcategory_panel = sliderule_wrapper.find("div", class_="mobile__menu__dropdown")
                if subcategory_panel:
                    for subcategory in subcategory_panel.find_all("a", class_="sliderow__title"):
                        subcategory_name = subcategory.get_text(strip=True)
                        subcategory_link = subcategory.get("href")
                        if subcategory_link:
                            subcategories.append({
                                "name": subcategory_name,
                                "link": f"https://sunnamusk.com{subcategory_link}" if subcategory_link.startswith("/") else subcategory_link
                            })

                # Add main category and its subcategories to the result
                categories[main_category_name] = subcategories

    return categories


```

```python
import requests
from bs4 import BeautifulSoup

def get_product_details(url):
    response = requests.get(url)
    response.raise_for_status()  # Raise an error if the request fails
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract Title
    title = soup.find('h1', class_='product__title').get_text(strip=True) if soup.find('h1', class_='product__title') else None

    # Extract Description
    description_tag = soup.find('div', class_='product__description__content__inner')
    description = description_tag.find('p').get_text(strip=True) if description_tag and description_tag.find('p') else None

    # Extract Fragrance Family
    fragrance_family_tag = soup.find('p', text="Fragrance Family")
    fragrance_family = fragrance_family_tag.find_next('p').get_text(strip=True) if fragrance_family_tag else None

    # Extract All Notes
    all_notes_tag = soup.find('p', text="All Notes")
    all_notes = all_notes_tag.find_next('div', class_='metafield-rich_text_field').get_text(strip=True) if all_notes_tag else None

    # Extract Gender
    gender_tag = soup.find('p', text="Gender")
    gender = gender_tag.find_next('p').get_text(strip=True) if gender_tag else None

    # Extract Price
    price = soup.find('div', {'data-wlh-price': True})['data-wlh-price'] if soup.find('div', {'data-wlh-price': True}) else None

    # Extract Detailed Notes (Top, Heart, Base)
    detailed_notes = {}
    for note in soup.find_all('div', class_='grid-item'):
        note_type = note.find('h3', class_='column__heading').get_text(strip=True) if note.find('h3', class_='column__heading') else None
        note_name = note.find('p', class_='column__subtitle').get_text(strip=True) if note.find('p', class_='column__subtitle') else None
        note_description = note.find('p', class_='column__text').get_text(strip=True) if note.find('p', class_='column__text') else None
        if note_type:
            detailed_notes[note_type] = {
                'name': note_name,
                'description': note_description
            }

    # Combine Results
    return {
        'Title': title,
        'Description': description,
        'Fragrance Family': fragrance_family,
        'All Notes': all_notes,
        'Gender': gender,
        'Price': price,
        'Detailed Notes': detailed_notes
    }
```

```python
def fetch_and_update_product_details(categories_dict):
    # Iterate over categories
    for category_key, category_data in categories_dict.items():
        for category in category_data:
            # Ensure the 'products' key exists
            if 'products' in category and isinstance(category['products'], list):
                updated_products = []
                for product_url in category['products']:
                    try:
                        # Validate product URL
                        if not isinstance(product_url, str) or not product_url.startswith('http'):
                            print(f"Invalid URL: {product_url}")
                            continue
                        
                        # Fetch product details for each valid URL
                        product_details = get_product_details(product_url)
                        updated_products.append({
                            'url': product_url,
                            'details': product_details
                        })
                    except Exception as e:
                        print(f"Error fetching details for {product_url}: {e}")
                        continue
                # Replace product URLs with detailed information
                category['products'] = updated_products
    return categories_dict
```

```python
def get_unique_titles(data):
    """
    Filters the products in the data to ensure unique titles.

    Args:
        data (dict): The original dictionary containing products.

    Returns:
        dict: The dictionary with unique titles.
    """
    seen_titles = set()
    unique_data = {}

    for category_key, category_list in data.items():
        unique_data[category_key] = []
        for category in category_list:
            unique_products = []
            for product in category.get('products', []):
                title = product.get('details', {}).get('Title')
                if title and title not in seen_titles:
                    seen_titles.add(title)
                    unique_products.append(product)
            category['products'] = unique_products
            unique_data[category_key].append(category)

    return unique_data
```

```python
def generate_markup(data):
    markup = ""

    for category in data.values():  # Loop through categories
        for item in category:  # Loop through items in each category
            category_name = item.get('name', 'Unknown Category')
            products = item.get('products', [])
            
            for product in products:  # Loop through products
                details = product.get('details', {})
                
                # Skip products without meaningful data
                if not details:
                    continue
                
                # Product title
                title = details.get('Title')
                if not title:
                    continue  # Skip if title is not available
                
                markup += f"# {title}\n"
                
                markup += f"## Name: {title}\n"
                # Category
                markup += f"## Category: {category_name}\n"
                
                # Description
                description = details.get('Description')
                if description:
                    markup += f"## Description:\n{description}\n"
                
                # Price
                price = details.get('Price')
                if price:
                    markup += f"## Price: {price}\n"
                
                # Additional Details
                fragrance_family = details.get('Fragrance Family')
                all_notes = details.get('All Notes')
                gender = details.get('Gender')
                detailed_notes = details.get('Detailed Notes', {})

                if any([fragrance_family, all_notes, gender]):
                    markup += "## Details:\n"
                    if fragrance_family:
                        markup += f"This fragrance belongs to the {fragrance_family} fragrance family. "
                    if all_notes:
                        markup += f"The scent features the following notes: {all_notes}. "
                    if gender:
                        markup += f"- Gender: {gender}\n"
                
                # Add detailed notes if available
                if detailed_notes:
                    for note_type, note in detailed_notes.items():
                        note_name = note.get('name')
                        note_description = note.get('description')
                        if note_name and note_description:
                            markup += f"- {note_type}: {note_name} - {note_description}\n"

                markup += "\n"

    return markup
```

```python

# Generate markup
markup_output = generate_markup(deatails_dict_unique)

# Save to file
output_file = "markup_output.txt"
with open(output_file, "w") as file:
    file.write(markup_output)

print(f"Markup saved to {output_file}")
```
